```python
import altair as alt
import pandas as pd

df= pd.read_csv("Absenteeism_at_work.csv")

df.head()
```

```python
 alt.Chart(df).mark_bar().encode(x = "Age", y = "Absenteeism time in hours")
```

```python
 alt.Chart(df).mark_bar().encode(x = "Age", y = "Absenteeism time in hours").interactive()
```

```python
brush = alt.selection_interval()
alt.Chart(df).mark_bar().encode(
    x = "Age", 
    y = "Absenteeism time in hours",
    color = alt.condition(brush, "Absents:N", alt.value("orange"))
).add_selection(
    brush
)
```
